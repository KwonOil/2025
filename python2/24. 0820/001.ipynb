```python
## 서울경제신문 스크래핑
# https://www.sedaily.com/NewsList/GD05
# 썸네일주소
# 제목
# 링크
# 요약내용

import requests
from bs4 import BeautifulSoup
import re
from datetime import datetime, timedelta

url = "https://www.sedaily.com/NewsList/GD05"

response = requests.get(url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, 'html.parser')

news_items = soup.select("ul.sub_news_list.type > li")

keywords = ['로봇','AI']

for i, item in enumerate(news_items, 1):
    title = item.select_one("div.article_tit").text
    desc = item.select_one("div.text_sub").text.strip()

    flow = False
    for keyword in keywords:
        if keyword not in desc: flow = True
    if not flow: continue

    img = item.select_one("div.thumb img")['src']
    link = item.select_one("div.article_tit > a")['href']
    tmp_date = item.select_one("div.text_info .date")
    tmp_time = item.select_one("div.text_info .time")
    tmp_reltime = item.select_one("div.text_info .rel_time")

    date = tmp_date.text.strip() if tmp_date else "날짜 없음"
    time = tmp_time.text.strip() if tmp_time else "시간 없음"
    reltime = tmp_reltime.text.strip() if tmp_reltime else date + " " + time

    if '분전' in reltime:
        rel = int(reltime.replace('분전',''))
        reltime = (datetime.now() - timedelta(minutes=rel)).strftime('%Y.%m.%d %H:%M:00')
    
    print("-"*50,str(i),"-"*50)
    print(f"썸네일: {img}")
    print(f"제목: {title}")
    print(f"링크: {link}")
    print(f"요약: {desc}")
    print(f"시간: {reltime}")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# url = "https://news.naver.com/breakingnews/section/103/241"
url = "https://news.naver.com/breakingnews/section/105/230"

response = requests.get(url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, 'html.parser')

news_items =  soup.select("div#newsct ul.sa_list > li.sa_item")

os.makedirs('img', exist_ok=True)

for item in news_items:
    thumb = item.select_one("div.sa_thumb a")
    if thumb:
        img = thumb.select_one('img').get('data-src','')
        filename = os.path.basename(img).split("?")[0]
        # print(filename)
        response = requests.get(img)
        with open('img/'+filename, 'wb') as f:
            f.write(response.content)
    else:
        img = '없음'
    
    title = item.select_one("div.sa_text a").text.strip()
    link = item.select_one("div.sa_text a").get('href','')
    info = item.select_one("div.sa_text_press").text.strip()
    abs = item.select_one("div.sa_text_lede").text.strip()

    response2 = requests.get(link)
    response2.encoding = 'utf-8'
    soup2 = BeautifulSoup(response2.text, 'html.parser')
    desc = soup2.select_one("article#dic_area").text[:150].replace("\n","").strip()

    print(f"썸네일: {img}")
    print(f"제목: {title}")
    print(f"링크: {link}")
    print(f"정보: {info}")
    print(f"요약: {abs}")
    print(f"내용: {desc}")

    print("-"*70)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re
u_agent = {
    'User-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36'
}

keyword = '맥북'
for num in range(0,10):
    url = f"https://www.clien.net/service/board/sold?&od=T31&category=0&po={num}"
    response = requests.get(url, headers = u_agent)
    response.encoding = 'utf-8'
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

    items = soup.select("div[data-role='list-title']")
    
    print("-"*50,str(num+1),"page","-"*50)
    for item in items:
        state = item.select_one("span.category,fixed").text.strip()
        if state == '판매':
            title = item.select_one("div.list_title span.subject_fixed").text.strip()
            link = item.select_one("div.list_title a.list_subject").get('href','')
            link = urljoin(url, link)
            if re.search(keyword,title):
                print(f"{state} | {title} | {link}")
```

```python
import requests

def check_robots_txt(url):
    """
    특정 웹사이트의 robots.txt 파일을 확인하는 함수
    Args:
        url (str): 웹사이트 URL
    """
    # robots.txt 파일의 URL 생성
    robots_url = f"{url}/robots.txt"

    try:
        # HTTP GET 요청으로 파일 가져오기
        response = requests.get(robots_url)

        # 요청 성공 시 상태 코드 및 내용 출력
        if response.status_code == 200:
            print(f"Status Code: {response.status_code} (Success)")
            print("robots.txt Content:")
            print(response.text)
        else:
            print(f"Status Code: {response.status_code} (No robots.txt found)")

    except Exception as e:
        print(f"Error occurred: {e}")

check_robots_txt("http://lecture.pul.kr")
```
