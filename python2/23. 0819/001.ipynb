```python
# 문서 제목, 내용 제목, 본문, 링크 목록
import requests
from bs4 import BeautifulSoup

url = "http://myoil.dothome.co.kr/ex1.html"

# with open("ex1.html", "r", encoding="utf-8") as f:
#     html = f.read()

response = requests.get(url)
response.encoding = 'utf-8'
html = response.text
soup = BeautifulSoup(html, 'html.parser')

links = soup.find_all('a')
imgs = soup.find(id = 'gallery').find_all('img')

print(f"문서 제목 : {soup.find('title').text}")
print(f"내용 제목 : {soup.find('h1').text}")
print(f"본문 : {soup.find('p', class_ = 'content').text}")
print("링크 목록")
for link in links:
    print(f"- {link['href']}")
print("이미지 목록")
for img in imgs:
    print(f"- {img['src']} | {img.get('alt', '없음')}")
soup
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://ssacademy.dothome.co.kr/003.html"

response = requests.get(url)
response.encoding = 'utf-8'
html = response.text
soup = BeautifulSoup(html, 'html.parser')

# 1. 모든 제목(h) 선택
h_tags = soup.select('h1, h2, h3, h4, h5, h6')
for i, h_tag in enumerate(h_tags, 1):
    print(f"{i}) {h_tag.text}")
print()

# 2. 모든 단락(p) 선택
p_tags = soup.select('p')
for i, p_tag in enumerate(p_tags, 1):
    print(f"{i}) {p_tag.text}")
print()

# 3. story 클래스를 가진 단락
story_class = soup.select('p.story')
for i, story_class in enumerate(story_class, 1):
    print(f"{i}) {story_class.text}")
print()

# 4. id가 ve-list 내에서 모든 목록 선택
ve_id = soup.select('#ve-list li')
for i, ve_id in enumerate(ve_id, 1):
    print(f"{i}) {ve_id.text}")
print()

# 5. 속성이 data-lo가 ko인 리스트 항목 선택
data_lo = soup.select('li[data-lo="ko"]')
for i, data_lo in enumerate(data_lo, 1):
    print(f"{i}) {data_lo.text}")
print()

lo = soup.select('li[data-lo]')
for i, lo in enumerate(lo, 1):
    print(f"{i}) {lo.text}")

soup
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = "http://lecture.pul.kr/ex2.html"

response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')

product_tag = soup.select('div.product')

for product in product_tag:
    name = product.select_one('.name').text
    price = int(product.select_one('.price').text.replace(',','').replace('원',''))
    description = product.select_one('.description').text
    link = product.select_one('a').get('href')
    link = urljoin(url, link)
    print(f"상품명 : {name}, 가격 : {price:,}원, 설명 : {description}, 링크 : {link}")

soup
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = "http://lecture.pul.kr/ex3.html"

response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')

product_tag = soup.select('div.product')

def choose_tags(product):
    name = product.select_one('.name').text
    price = int(product.select_one('.price').text.replace(',','').replace('원',''))
    description = product.select_one('.description').text
    link = product.select_one('.description > a').get('href','')
    link = urljoin(url, link)

    return name, price, description, link

keys = ['상품명', '가격', '설명', '링크']
result = [dict(zip(keys, choose_tags(product))) for product in product_tag]

print("[미션1]\n")
for product in result:
    p = list(product.items())
    for i in range(len(p)):
        print(f"{p[i][0]} : {p[i][1]:<12}\t|", end = '')
    print()

print("[미션2]")
result
```

```python
import requests
from bs4 import BeautifulSoup
import re
from urllib.parse import urljoin

url = "https://www.sedaily.com/NewsList/GD05"

response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')

ul = soup.select('ul.sub_news_list.type > li')

keyword = 'AI|로봇'

def choose_tags(ogj):
    thumb_link = ogj.select_one('div.thumb img').get('src','')
    title = ogj.select_one('div.article_tit > a').text
    link = ogj.select_one('div.article_tit > a')['href']
    text_sub = ogj.select_one('div.text_sub').text.strip()

    return thumb_link, title, link, text_sub

keys = ['썸네일','제목','링크','요약']
result = [dict(zip(keys,choose_tags(li))) for li in ul if re.search(keyword, li.select_one('div.text_sub').text.strip())]

for i, res in enumerate(result,1):
    print("-" * 10 + f" {i} " + "-" * 10)
    for k,v in res.items():
        print(f"{k:^4} : {v}")
    print()

```

```python
import requests
from bs4 import BeautifulSoup

url = "https://news.naver.com/breakingnews/section/103/241"

response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')

div = soup.select('div.section_latest_article li')

def choose_tags(ogj):
    return (
        ogj.select_one('div.sa_thumb_inner img')['data-src'], # 썸네일
        ogj.select_one('div.sa_text > a').text.strip(), # 제목
        ogj.select_one('div.sa_text > a')['href'], # 링크
        ogj.select_one('div.sa_text_lede').text.strip(), # 내용
        ogj.select_one('div.sa_text_press').text.strip(), # 신문사
    )

keys = ['썸네일', '제목', '링크', '내용', '신문사']
result = [dict(zip(keys,choose_tags(li))) for li in div]

for i, res in enumerate(result,1):
    print("-" * 10 + f" {i} " + "-" * 10)
    for k,v in res.items():
        print(f"{k:^4} : {v}")
    print()
```
