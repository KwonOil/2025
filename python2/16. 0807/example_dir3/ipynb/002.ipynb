```python
# 백업
# 1) 백업폴더가 존재하는지
# 2) 백업폴더가 없으면 생성 아니면 계속 진행
# 3) 원본폴더의 파일목록 불러오기
# 4) 파일목록 순서대로 돌기
# 5) 전체경로 만들기(파일)
# 6) 목록이 파일인지 확인 (파일일 경우에만 백업진행)
# 7) 작업폴더의 파일을 백업폴더의 파일에 저장
import os
import shutil
source_directory = os.getcwd()
backup_directory = os.getcwd() + '/example_dir3'
def backup_files(src,dst):
    # 1)
    if not os.path.exists(dst):
        # 2)
        os.makedirs(dst, exist_ok = True)
    
    # 4)
    for filename in os.listdir(src):
        # 3)
        # print(filename)
        # 5)
        source_path = os.path.join(src,filename)
        backup_path = os.path.join(dst,filename)

        # 6)
        if os.path.isfile(source_path):
            # 7)
            shutil.copy(source_path, backup_path)
            if os.path.exists(backup_path):
                print(f"{backup_path} - 백업 완료")
            else:
                print(f"{backup_path} - 백업 실패")
    


backup_files(source_directory, backup_directory)
```
