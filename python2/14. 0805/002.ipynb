```python
from datetime import datetime, timedelta
import time

def start_timer(minutes):
    """타이머를 시작하는 함수"""
    duration = timedelta(minutes=minutes)  # 타이머 시간 설정
    end_time = datetime.now() + duration  # 종료 시간 계산

    print(f"타이머 시작: {minutes}분 후에 알림이 옵니다.")
    while datetime.now() < end_time:
        time_left = end_time - datetime.now()

        # 같은 줄에 남은 시간 출력
        print(f"남은 시간: {time_left}", end='\r')
        time.sleep(1)  # 1초 대기

    print("\n타이머 완료! 시간입니다!")

# 사용자로부터 분 단위 입력 받기
try:
    minutes = int(input("타이머를 몇 분으로 설정하시겠습니까? "))
    if minutes <= 0:
        raise ValueError("0 또는 그 이하의 값은 입력할 수 없습니다.")
    start_timer(minutes)
except ValueError as e:
    print(f"잘못된 입력: {e}")
```

```python
from datetime import datetime

# 일정 --> (제목 : 년월일시분)
# [('회의','2025-05-20 15:00'),('저녁약속','2025-05-24 19:30')]
events = []
now = datetime.now().strftime("%Y-%m-%d %H:%M")
def add_event(name, edate):
    edate2 = edate.strftime("%Y-%m-%d %H:%M")
    events.append((name, edate2))

def show_events():
    if not events:
        print("일정이 없습니다")
        return
    print("[모든 일정]")
    for event_name, event_date in events:
        print(f"- {event_name} : {datetime.strptime(event_date,'%Y-%m-%d %H:%M').strftime('%Y년 %m월 %d일 %H시 %M분')}")
        #print(f"- {event_name} : {event_date.strftime('%Y년 %m월 %d일 %H시 %M분')}")

def remain_events():
    if not events:
        print("일정이 없습니다")
        return

    print("[남은 일정]")
    for event_name, event_date in events:
        if now <= event_date:
            print(f"- {event_name} : {datetime.strptime(event_date,'%Y-%m-%d %H:%M').strftime('%Y년 %m월 %d일 %H시 %M분')}")


add_event("회의", datetime(2025,5,10,15,0))
add_event("생일", datetime(2025,12,16,20,0))
add_event("저녁", datetime(2025,8,8,19,30))

print(events)
print()
show_events()
print()
remain_events()
```

```python
from datetime import datetime
import os

events = []
now = datetime.now().strftime('%Y-%m-%d %H:%M')

def clear_screen():
    os.system('cls') if os.name == 'nt' else os.system('clear')

def add_event(name, edate):
    edate2 = edate.strftime('%Y-%m-%d %H:%M')
    events.append([name, edate2])
    save_event(events)

def save_event():
    with open("events.txt", 'w', encoding = 'utf-8') as f:
        for event_name, event_date in events:
            f.write(f"{event_name}|{event_date}\n")

def load_event():
    with open("events.txt","r", encoding='utf-8') as f:
        for line in f:
            event_name, event_date = line.strip().split('|')
            events.append([event_name, event_date])

def show_events():
    if not events:
        print("일정이 없습니다.")
        return

    print("\n[모든 일정]")
    for event_name, event_date in events:
        print(f"- {event_name} : {datetime.strptime(event_date,'%Y-%m-%d %H:%M').strftime('%Y년 %m월 %d일 %H시 %M분')}")

def remain_events():
    if not events:
        print("일정이 없습니다.")
        return

    print("\n[남은 일정]")
    for event_name, event_date in events:
        if now <= event_date:
            print(f"- {event_name} : {datetime.strptime(event_date,'%Y-%m-%d %H:%M').strftime('%Y년 %m월 %d일 %H시 %M분')}")

def show_menu():
    print("\n일정 관리 프로그램")
    print("1. 일정 추가")
    print("2. 모든 일정 보기")
    print("3. 남은 일정 보기")
    print("4. 종료")

def main():
    load_event()
    while(True):
        clear_screen()
        show_menu()
        choice = input("원하는 작업을 선택하세요(1 ~ 4)")
        if choice == '4':
            break
        elif choice == '1':
            name = input("일정 제목을 입력하세요 :")
            edate = input("일정 날짜를 입력하세요(YYYY-MM-DD HH:MM) :")
            add_event(name, datetime.strptime(edate,'%Y-%m-%d %H:%M'))
        elif choice == '2':
            show_events()
        elif choice == '3':
            remain_events()
        input("계속하려면 아무 키나 누르세요...")

if __name__ == '__main__':
    main()
```
