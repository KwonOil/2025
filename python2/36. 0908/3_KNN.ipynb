# KNN(K-Nearest Neighbors)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier

df4 = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df4.info()
df4
```

```python
df4.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# 비권장(새로운 값 들어올 시 문제 발생)
# df4['sepal_length'] = scaler.fit_transform(df4[['sepal_length']])
# df4['sepal_width'] = scaler.fit_transform(df4[['sepal_width']])
# df4['petal_length'] = scaler.fit_transform(df4[['petal_length']])
# df4['petal_width'] = scaler.fit_transform(df4[['petal_width']])

# 권장
df4[['sepal_length','sepal_width','petal_length','petal_width']] = scaler.fit_transform(df4[['sepal_length','sepal_width','petal_length','petal_width']])
df4
```

```python
# 독립변수(피처), 종속변수(타겟) 지정
X3 = df4[['sepal_length','sepal_width','petal_length','petal_width']]
y3 = df4['species']
```

```python
# 데이터셋 분리 (학습 : 테스트 = 8 : 2)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3,y3,test_size = 0.2, random_state = 11)
print(X3_train.shape)
print(X3_test.shape)
print(y3_train.shape)
print(y3_test.shape)
```

```python
# 학습
model = KNeighborsClassifier(n_neighbors=9)
model.fit(X3_train, y3_train)
```

```python
# 예측
pred3 = model.predict(X3_test)
print('실제값:',y3_test.values)
print('예측값:',pred3)
```

```python
# 성능확인
rep = classification_report(y3_test,pred3)
mat = confusion_matrix(y3_test,pred3)

print(rep)
print(mat)
```
