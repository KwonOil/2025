```python
a = ['a','b','c']

# a 값 출력
for v in a:
    print(v)

# a 인덱스 출력
for i in range(len(a)):
    print(i)

# enumerate 사용
for i, v in enumerate(a):
    print(i,v)

g = ['a','b']
k = {}
for i, v in enumerate(a):
    print(1,v)
    k[i] = v
print(k)
print(k[0],g[0])
```

### 컴프리헨션(Comprehension)

```python
mylist = [i+1 for i in range(10)]
# [expression for item in iterable if condition]
mylist2 = [i**2 for i in mylist if i % 2 == 0]
mylist3 = [i**2 if i % 2 == 0 else 0 for i in mylist]
print(mylist)
print(mylist2)
print(mylist3)
```

```python
squares = [x**2 for x in range(10)]
squares2 = [x**2 for x in range(10) if x % 2 == 0]
print(squares)
print(squares2)
```

```python
txt = "Hello"
chars = [i.upper() for i in txt]
chars
```

```python
# 중첩 컴프리헨션
matrix = [[1,2,3],[4,5,6],[7,8,9]]
matrix2 = [n for m in matrix for n in m if n <= 7]
print(matrix2)
```

```python
# 딕셔너리 컴프리헨션
a = {x:x**2 for x in range(50) if x % 2 == 0}
print(a)
a[22]
```

```python
from attr import asdict


d = {
    "kim" : 35,
    "lee" : 30,
    "park" : 28
    }
e = {i:v for i,v in d.items() if i == 'kim' and v >= 30}
print(e)
```

```python
won = 1387
price = {"우유" : 1.5, "커피" : 4.5, "빵" : 2.5}
price2 = {k:v*won for k,v in price.items()}
print(price2)
```

```python
# kg ---> g
wk = {"사과" : 1.2, "수박" : 3.5, "포도" : 0.5}
print({k:v*1000 for k,v in wk.items()})
```

```python
# 20점 만점 ---> 100점 만점
s = {"Alive" : 18, "Bob" : 14, "Charlie" : 20}
print({k:v*5 for k,v in s.items()})
```

```python
# 3.3% 공제 후 급여
rate = 0.033
salary = {"Tom" : 4000, "Jane" : 5500, "Mike" : 6200}
print({k:v*(1-rate) for k,v in salary.items()})
```

```python
# 총액계산(수량,단가)
items = {"펜":(2,500),"노트":(3,2000),"USB":(1,8000)}
s4 = {k:v[0]*v[1] for k,v in items.items()}
print(s4)
print(f"총액 : {sum(s4.values()):,}원")
```

```python
# VIP 고객에게 10% 추가
points = {"user1" : 1200, "VIP_user2" : 3000, "VIP_user3" : 5000}
print({k:round(v*1.1) if 'VIP' in k else v for k,v in points.items()})
```
