```python
traffic_tuples = [
    ("2024-03-15",1500, 4500, 185, 25.5),
    ("2024-03-14",1400, 4200, 190, 23.8),
    ("2024-03-13",1600, 4800, 178, 26.2)
]

keys = ["날짜","방문자수","페이지뷰","평균체류시간","이탈률"]

mydic = [dict(zip(keys, item)) for item in traffic_tuples]

mydic
```

```python
search_date = input("조회할 날짜는(YYYY-MM-DD)?")

for item in mydic:
    if item['날짜'] == search_date:
        print(f"방문자 수 : {item['방문자수']}명")
        print(f"페이지 뷰  : {item['페이지뷰']}명")
        print(f"평균 체류시간 : {item['평균체류시간']}시간")
        print(f"이탈률 : {item['이탈률']}%")
```

```python
store_sales_tuples = [
    ("S001", "서울", 5000000, 200, 25000, 8),
    ("S002", "부산", 3500000, 150, 23000, 6),
    ("S003", "대구", 2800000, 120, 23500, 5)
]

# 매장코드, 지역, 일일판매액, 방문객수, 평균구매액, 직원수
keys = ["매장코드", "지역", "일일판매액", "방문객수", "평균구매액", "직원수"]
store_sales_dict = [dict(zip(keys, item)) for item in store_sales_tuples]

store_sales_dict
```

```python
shcode = input("조회할 지역은?")

for item in store_sales_dict:
    if item['지역'] == shcode:
        for j in range(2,len(keys)):
            print(f"{keys[j]}\n - {item[keys[j]]:,}")
```

```python
customer_dict = {
    "C001" : {"이름" : "김철수", "나이" : 35, "성별" : "남성", "구매금액" : 150000, "구매날짜" : "2024-03-15"},
    "C002" : {"이름" : "이영희", "나이" : 28, "성별" : "여성", "구매금액" : 280000, "구매날짜" : "2024-03-14"},
    "C003" : {"이름" : "박민수", "나이" : 42, "성별" : "남성", "구매금액" : 95000, "구매날짜" : "2024-03-13"}
}

# 첫 번째 방법 - 초기화 시 키를 생성하고 시작하는 방법
# customers = {'남성' : [], '여성' : []}
# for info in customer_dict.values():
#     customers[info["성별"]].append(info["구매금액"])


# 두 번째 방법 - 빈 객체로 초기화한 후 값이 없을 시 초기화하는 방법
customers = {}

for info in customer_dict.values():
    if info["성별"] not in customers:
        customers[info["성별"]] = []
    customers[info["성별"]].append(info["구매금액"])

print(f"- {customers}")

def cal_avg(li):
    return sum(li)/len(li)

for v in customers.keys():
    print(f"- {v}의 구매평균 : {cal_avg(customers[v]):,.0f}원입니다")
```
