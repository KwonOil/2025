```python
import datetime
import os

def file_info(file_path):
    # 바이트단위 사이즈
    file_size = os.path.getsize(file_path)

    # 파일시간 확인 (초단위, 유닉스 시간)
    # getctime : 생성 시간 (Creation Time)
    #times = os.path.getctime(file_path)
    # getmtime : 수정 시간 (Modified Time)
    times = os.path.getmtime(file_path)
    # getatime : 접근 시간 (Access Time)
    #times = os.path.getatime(file_path)

    date_obj = datetime.datetime.fromtimestamp(times)
    date_str = date_obj.strftime('%Y-%m-%d %H:%M:%S')

    print(f"파일 크기: {file_size} bytes")
    print(f"최종 수정 날짜: {date_str}")

file_info("001.ipynb")
```

```python
import os

def flist(source_directory, n):
    # source_directory 내 목록(파일, 디렉토리)
    for filename in os.listdir(source_directory):
        #continue
        source_path = os.path.join(source_directory, filename)
        
        if os.path.isfile(source_path):
            print("\t"*(n-1), end="")
            print("   ▷", filename, "(", os.path.getsize(source_path), " bytes)")
            pass
        elif os.path.isdir(source_path):
            print("\t"*n, end="")
            print("■ [", n+1, ']', source_path)

            flist(source_path, n+1)

flist(r'C:\eclipse', 0)
```

```python
### 특정 폴더 내 모든 파일크기 및 합산크기 1
import os
directory = r"C:/Temp/2025/python2/16. 0807/example_dir3"

def total_filesize(directory):
    total = 0
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        if os.path.isfile(file_path):
            fsize = os.path.getsize(file_path)
            print(f"{filename} : {fsize} bytes")
            total += fsize
        else:
            total += total_filesize(file_path)
    return total

totalsize = total_filesize(directory)
print(f"전체 크기 : {totalsize} bytes")
```

```python
### 특정 폴더 내 모든 파일크기 및 합산크기 2
import os
directory = r"C:/Temp/2025/python2/16. 0807/example_dir3"

def total_filesize2(directory):
    total = 0
    for dirpath, dirnames, filenames in os.walk(directory):
        ## dirpath : 모든 디렉토리 정보(하위 포함)
        ## dirname : 디렉토리 내 디렉토리 목록(리스트)
        ## filenames : 디렉토리 내 파일 목록(리스트)
        # print(dirpath, dirname, filenames)
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            if os.path.isfile(file_path):
                fsize = os.path.getsize(file_path)
                print(f"{filename} : {fsize} bytes")
                total += fsize
    return total

totalsize2 = total_filesize2(directory)
print(f"전체 크기 : {totalsize2} bytes")
```

```python
### 특정 폴더 내의 특정 확장자를 가진 파일 일괄 삭제
import os
directory = r"C:/Temp/2025/python2/16. 0807/example_dir3"

def total_filesize3(directory, target_ext):
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            ext = os.path.splitext(filename)[1].split('.')
            # print(ext)
            if ext == target_ext:
                os.remove(file_path)
                if not os.path.exists(file_path):
                    print(f"{file_path} - 삭제 완료")
total_filesize3(directory, 'txt')
```

```python
### 생성된지 10일이 지난 파일들 지우기
import os
from datetime import datetime, timedelta

directory = r"C:/Temp/2025/python2/17. 0808/sample"

def total_filesize4(directory, days):
    times = datetime.now() - timedelta(days=days)
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            creation_time = datetime.fromtimestamp(os.path.getctime(file_path))
            # print(time.time() - times, 60 * 60 * 24 * 10)
            if creation_time < times:
                os.remove(file_path)
                if not os.path.exists(file_path):
                    print(f"{file_path} - 삭제 완료")
totalsize4 = total_filesize4(directory, 10)
```

```python
import os
from datetime import datetime, timedelta

def delete_old_files(directory, days):
    now = datetime.now()
    print(f"now: {now}")

    # 현재시간에 days 날짜 만큼 빼기
    tendays_ago = now - timedelta(days=days)
    print(f"tendays_ago: {tendays_ago}")

    # 디렉토리 내 모든 파일 확인
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        # 파일인지 확인
        if os.path.isfile(file_path):
            unix_m_timestamp = os.path.getmtime(file_path)
            
            # fromtimestamp: 유닉스타임을 년월일시분초 형태의 datetime 객체로 변환
            date_obj = datetime.fromtimestamp(unix_m_timestamp)
            print(f"date_obj: {date_obj}-{filename}")

            if date_obj < tendays_ago:
                os.remove(file_path)

directory_path = r'C:\Temp\2025\python2\17. 0808\sample'
delete_old_files(directory_path, 10)
```

```python
### 특정 폴더 내 모든 파일 이름에 '이름_'접두사를 추가하여 이름을 변경
import os

def raname_files_with_prefix(directory, prefix):
    for filename in os.listdir(directory):
        source_path = os.path.join(directory,filename)
        if os.path.isfile(source_path):
            new_filename = prefix + filename
            ## 바뀔 이름(경로)
            new_source_path = os.path.join(directory, new_filename)
            # 이름 변경
            os.rename(source_path, new_source_path)
            print(f"파일이 변경되었습니다 - {filename} -> {new_filename}")

deirectory = r"C:\Temp\2025\python2\17. 0808\sample"
raname_files_with_prefix(directory, "new_")
```

```python
### 특정 폴더 내 모든 파일 이름에 '이름_'접두사를 추가하여 이름을 변경
import os
from datetime import datetime

def raname_files_with_suffix(directory):
    for filename in os.listdir(directory):
        source_path = os.path.join(directory,filename)
        if os.path.isfile(source_path):
            datestr = datetime.now().strftime("_%Y-%m-%d_%H-%M-%S")
            name, ext = filename.split('.')
            new_filename = name + datestr + "." + ext
            ## 바뀔 이름(경로)
            new_source_path = os.path.join(directory, new_filename)
            # 이름 변경
            os.rename(source_path, new_source_path)
            print(f"파일이 변경되었습니다 - {filename} -> {new_filename}")

deirectory = r"C:\Temp\2025\python2\17. 0808\sample"
raname_files_with_suffix(directory)
```

```python
def list_sorted_files_by_size(target_directory):
    files = os.listdir(target_directory)
    files_with_size = [(filename, os.path.getsize(os.path.join(target_directory,filename))) for filename in files]

    sorted_files = sorted(files_with_size, key = lambda x : x[1], reverse = True)

    for filename, size in sorted_files:
        print(f"{filename} : {size:,} bytes")

target_directory = r"C:\Temp\2025\python2\17. 0808"
list_sorted_files_by_size(target_directory)
```
