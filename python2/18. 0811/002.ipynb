```python
# 지정 폴더에서 특정 확장자를 가진 파일만 골라서 압축

import os
import shutil

target_dir = './logs'
temp_dir = './temp_logs_to_zip'
zip_path = 'logs_backup'

# 1. 임시 폴더 생성 및 특정 확장자 파일 복사
os.makedirs(temp_dir, exist_ok=True)
for filename in os.listdir(target_dir):
    path = os.path.join(target_dir, filename)
    if filename.endswith('.log'):
        shutil.copy2(path, temp_dir)

# 2. 임시 폴더 전체를 ZIP 압축
shutil.make_archive(zip_path, 'zip', temp_dir)

# 3. 임시 폴더 삭제
shutil.rmtree(temp_dir)

print(f"로그 파일 ZIP 압축 완료: {zip_path}.zip")
```

### Class
- 클래스는 객체 지향 프로그래밍(OOP(Object-Oriented Programming))에서 객체를 정의하는 구조
- 속성과 메서드를 포함하여 데이터와 기능을 함께 묶어 처리할 수 있도록 합니다.
- 클래스는 코드의 재사용성과 구조적 조직을 돕습니다.

```python
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))

print(add2(5))
print(add2(7))
```

```python
# 클래스 정의 (속성 + 메소드)

class Calculate:

    # 생성자
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result

# 객체 생성
cal1 = Calculate()
cal2 = Calculate()

print(cal1.result)
print(cal1.add(3), cal1.result)
print(cal1.add(5), cal1.result)
print(cal1.add(7), cal1.result)

print("="*50)

print(cal2.result)
print(cal2.add(2), cal2.result)
print(cal2.add(4), cal2.result)
print(cal2.add(6), cal2.result)
print(cal2.sub(2), cal2.result)

```

```python
class Dog:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name}가 멍멍 합니다."

    def get_age(self):
        return f"{self.name}의 나이는 {self.age}세 입니다."
    
dog1 = Dog('버디', 7)
dog2 = Dog('신디', 5)

print(dog1.name, dog1.age)
print(dog1.bark(), dog1.get_age())
print(dog2.bark(), dog2.get_age())
```

```python
class FourCal:

    def __init__(self, first=0, second=0):
        self.first = first
        self.second = second

    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def div(self):
        result = self.first / self.second
        return result
    
#a = FourCal()
# a = FourCal(7, 8)
# print(a.first, a.second)

# a.setdata(8, 9)
# print(a.first, a.second)

a = FourCal(4, 2)
b = FourCal(8, 4)

print(a.add(), a.sub(), a.mul(), a.div(), sep=", ")
print(b.add(), b.sub(), b.mul(), b.div(), sep=", ")
```

```python
# 1. 상속
class AddFourCal(FourCal):
    
    def pow(self):
        result = self.first ** self.second
        return result

c = AddFourCal(5, 3)
print(c.add(), c.pow())
```

```python
d = FourCal(4, 0)
print(d.add())
print(d.sub())
print(d.mul())
print(d.div())
```

```python
# 2. 다형성: 메소드 오버라이드
class SafeFourCal(FourCal):

    def div(self):
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
        
e = SafeFourCal(3, 0)

print(e.add())
print(e.sub())
print(e.mul())
print(e.div())
```

```python
# 4. 추상화
# - 코드의 표준화와 일관성 유지: 추상 메소드로 자식 클래스들이 반드시 구현해야 하는 메소드를 정해두어, 다양한 클래스가 같은 인터페이스(메소드 구조)를 따르도록 한다.
# - 메소드 구현 강제: 자식 클래스가 특정 메소드를 반드시 구현하도록 만들어 버그나 누락 위험을 줄임
from abc import ABC, abstractmethod

# 도형
class Shape(ABC):
    
    @abstractmethod
    def area(self): # 면적
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height  

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height  

circle = Circle(5)
print(circle.area())

square = Square(4)
print(square.area())

rectangle = Rectangle(4, 6)
print(rectangle.area())

triangle = Triangle(3, 6)
print(triangle.area())
```

```python
#make_sound와 get_info는 추상 메서드이기 때문에 Animal 클래스를 상속받는 모든 서브클래스에서 반드시 구현해야 합니다.

from abc import ABC, abstractmethod

# 추상 클래스
class Animal(ABC):

    @abstractmethod
    def make_sound(self):
        pass

    @abstractmethod
    def get_info(self):
        pass


# 올바르게 구현한 서브클래스
class Dog(Animal):

    def make_sound(self):
        return "멍멍!"

    def get_info(self):
        return "나는 강아지입니다."


# 구현이 일부 빠진 서브클래스 -> 오류 발생
class Cat(Animal):

    def make_sound(self):
        return "야옹~"

    # get_info 구현 누락
    def get_info(self):
        return "나는 고양이입니다."


# 실행 예제
if __name__ == "__main__":
    dog = Dog()
    print(dog.make_sound())  # "멍멍!"
    print(dog.get_info())    # "나는 개입니다."

    # 추상 메서드를 모두 구현하지 않은 Cat 클래스를 인스턴스화하면 에러 발생
    try:
        cat = Cat()  # TypeError 발생
        print(cat.make_sound())
        print(cat.get_info())
    except TypeError as e:
        print("에러:", e)
```

```python
# 다중 상속
class Mother:
    def mother_method(self):
        print("어머니 클래스 메소드")

class Father:
    def father_method(self):
        print("아버지 클래스 메소드")

class Child(Mother, Father):
    def child_method(self):
        print("자식 클래스 메소드")

child_instance = Child()
child_instance.mother_method()  # 어머니 클래스 메소드
child_instance.father_method()  # 아버지 클래스 메소드
child_instance.child_method()   # 자식 클래스 메소드
```

```python
# Student("홍길동", {"국어": 85, "영어": 90, "수학": 78})

class Student:
    
    # scores: 과목별 성적데이터는 딕셔너리로 구성 
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores

    def total_score(self):
        return sum(self.scores.values())

    def avg_score(self):
        return self.total_score() / len(self.scores)

    def show_info(self):
        print(f"학생 이름: {self.name}")
        for subject, score in self.scores.items():
            print(f"{subject}: {score}")
        print(f"총점: {self.total_score()}, 평균: {self.avg_score():.2f}\n")

std1 = Student("홍길동", {"국어": 85, "영어": 90, "수학": 78})
std2 = Student("김영희", {"국어": 92, "영어": 88, "수학": 95})
std3 = Student("이철수", {"국어": 70, "영어": 75, "수학": 80})
#std1.show_info()
#print(std2.show_info())
#print(std3.show_info())

students = []

students.append(Student("이순신", {"국어": 85, "영어": 90, "수학": 78}))
students.append(Student("김영희", {"국어": 92, "영어": 88, "수학": 95}))
students.append(Student("이철수", {"국어": 70, "영어": 75, "수학": 80}))


for student in students:
    student.show_info()
```
