```python
username = input("사용자 아이디를 입력하세요")
password = input("비밀번호를 입력하세요:")
uid = "admin"
upw = "1234"

if(username == uid and password == upw):
    print("관리자님 로그인되었습니다")
elif(username == uid and password != upw):
    print("비밀번호를 잘못 입력했습니다")
elif(username != uid):
    print("아이디를 잘못 입력하셨습니다")
else:
    print("???")
```

```python
length = 20
for i in range(1,length+1):
    bnk = " " * (length - i)
    star = '*' * (2*i-1)
    print((bnk + star + bnk))
```

```python
# length에서 0까지 줄여가면서 실행
length = 20
for i in range(length,0,-1):
    bnk = "0" * (length - i)
    star = '*' * (2*i-1)
    print((bnk + star + bnk))
```

```python
# 0부터 length까지 실행
length = 20
for i in range(1, length):
    bnk = "0" * (i-1)
    star = '*' * (length*2 - 2*i-1)
    print((bnk + star + bnk))
```
