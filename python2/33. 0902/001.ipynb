```python
import pandas as pd
import requests
from io import StringIO

url = "https://finance.naver.com/sise/sise_market_sum.naver"

headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36'
}
response = requests.get(url, headers = headers)
response.encoding = 'euc-kr'
html_io = StringIO(response.text)

tables = pd.read_html(html_io, encoding = 'euc-kr')
df = tables[1]
# df.dropna = 결측치 제거 함수
# 기본 축: axis=0이 기본 → “행”을 검사해 제거
# 기준: how='all' → 전부 결측치일 때만 제거
# 대조: how='any'는 하나라도 결측치가 있으면 제거
# 부분열만 검사: subset=['열A','열B']로 검사 대상 열 한정 가능
# 열 제거: axis=1을 주면 “열” 기준으로 동작
# thresh와 동시 사용 금지: how와 thresh는 함께 쓸 수 없습니다(둘 다 주면 에러)
df = df.dropna(how = 'all')             # 모든 행이 NaN일 때
df = df.dropna(how = 'all', axis = 1)   # 모든 열이 NaN일 때
df.head(10)
```

#### 판다스
지원 파일 형식 : csv,excel, html, json, sql

```python
import pandas as pd
from sqlalchemy import create_engine

data = {
    '이름' : ['김철수', '이영희', '박민수', '정성훈'],
    '나이' : [25, 30, 20, 35],
    '성적' : [85.5, 92.3, 78.8, 96.5],
    '합격여부' : [True, True, False, True]
}
# df 만들기
df = pd.DataFrame(data)

# 특정 형식으로 저장하기
# to csv
df.to_csv('df_to_csv.csv', encoding='utf-8', index = False)
# to excel
df.to_excel('df_to_excel.xlsx', sheet_name='ex', index=False)
# to html
with open('df_to_html.html','w', encoding='utf-8') as f:
    df.to_html(f, index = False)
# to json
df.to_json('df_to_json.json', orient = 'records', force_ascii = False)
# to db(sqlite)
engine = create_engine('sqlite:///df_to_sql.db')
df.to_sql('dftable', engine, if_exists='replace', index = False)

# 저장된 데이터 파일 불러오기
# load csv
df_csv = pd.read_csv('df_to_csv.csv')
print('[csv]\n',df_csv)
# load excel
df_excel = pd.read_excel('df_to_excel.xlsx')
print('\n[excel]\n',df_excel)
# load html
df_html = pd.read_html('df_to_html.html', encoding='utf-8')[0]
print('\n[html]\n',df_html)
# load json
df_json = pd.read_json('df_to_json.json')
print('\n[json]\n',df_json)
# load sql
sql = '''
    select * from dftable
'''
df_db = pd.read_sql(sql, engine)
print('\n[sql]\n',df_db)
```
