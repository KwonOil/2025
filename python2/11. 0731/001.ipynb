```python
# 미션 1: 텍스트 파일 읽기 및 쓰기
# 목표: 사용자로부터 이름과 이메일을 입력받아 텍스트 파일에 저장하기.
# 단계:
# 사용자로부터 이름과 이메일을 입력받습니다.
# 입력받은 데이터를 data.txt 파일에 저장합니다.
# 파일을 다시 읽어 저장된 내용을 출력합니다.

import os

def write_file4csv(filename):
    name = input("이름을 입력하세요")
    email = input("이메일을 입력하세요")
    
    if not os.path.exists(filename) or os.path.getsize(filename) == 0:
        with open(filename, "w", encoding = 'utf-8') as f:
            f.write("이름, 이메일\n")

    with open(filename, "a+", encoding = 'utf-8') as f:
        if ',' in name: name = f'"{name}"'
        if ',' in email: email = f'"{email}"'
        f.write(f'''{name}, {email}\n''')

def read_file4csv(filename):
    with open(filename, "r", encoding = 'utf-8') as f:
        content = f.read()
    return content

fname = "data_txt.csv"
write_file4csv(fname)
cont = read_file4csv(fname)
print(cont)
```

```python
# 미션 2: JSON 파일 생성 및 수정
# 목표: 학생 정보를 담은 JSON 파일 생성하기.
# 단계:
# - 학생의 이름, 나이, 전공을 입력받습니다.
# - 해당 데이터를 JSON 형식으로 변환하여 students.json 파일에 저장합니다.
# - JSON 파일을 읽어 전체 학생 목록을 출력합니다.

import json

def create_json(filename):
    header = ['이름','나이','전공']
    values = [input("이름을 입력하세요"), int(input("나이를 입력하세요")), input("전공을 입력하세요")]

    student = {h : values[i] for i, h in enumerate(header)}

    students = []
    students.append(student)

    with open(filename, 'w', encoding = 'utf-8') as f:
        json.dump(students, f, ensure_ascii=False, indent = 4)

def read_json(filename):
    with open('students.json', 'r', encoding = 'utf-8') as f:
        s = json.load(f)
    return s

fname = 'students.json'
# create_json(fname)
info = read_json(fname)

for i in info:
    print(i['이름'],i['나이'],i['전공'])
```

```python
# 미션 3: JSON 데이터 업데이트
# 목표: 기존의 JSON 파일에 새로운 학생 정보를 추가하기.
# 단계:
# students.json 파일을 열어 현재 학생 정보를 읽습니다.
# 추가할 학생 정보를 입력받아 기존 리스트에 추가합니다.
# 업데이트된 리스트를 다시 JSON 파일에 저장합니다.

import json

try:
    with open('students3.json', 'r', encoding = 'utf-8') as f:
        students = json.load(f)
except FileNotFoundError:
    students = []
        
header = ['이름','나이','전공']
values = [input("이름을 입력하세요"), input("나이를 입력하세요"), input("전공을 입력하세요")]
student = {}

for i, h in enumerate(header):
    student[h] = values[i]

students.append(student)


with open('students3.json', 'w', encoding = 'utf-8') as f:
    json.dump(students,f)

with open('students3.json', 'r', encoding = 'utf-8') as f:
    s = json.load(f)
print(s)
```

```python
import json

def update_json_file(filename):
    name = input("추가할 학생 이름을 입력하세요: ")
    age = int(input("학생 나이를 입력하세요: "))
    major = input("전공을 입력하세요: ")
    
    try:
        with open(filename, "r", encoding='utf-8') as file:
            students = json.load(file)
    except FileNotFoundError:
        students = []

    students.append({"name": name, "age": age, "major": major})

    with open(filename, "w", encoding='utf-8') as file:
        json.dump(students, file, ensure_ascii=False, indent=4)

update_json_file("students3.json")
```

```python
# 미션 4: JSON 데이터 검색
# 목표: 특정 학생 정보를 검색하기.
# 단계:
# students.json 파일을 읽어 모든 학생 정보를 가져옵니다.
# 검색할 학생의 이름을 입력받고, 해당 학생의 정보를 출력합니다.
# 학생이 존재하지 않을 경우 적절한 메시지를 출력합니다.

c_name = input("검색할 이름을 입력하세요")

try:
    with open('students.json', 'r', encoding = 'utf-8') as f:
        students = json.load(f)
        f = False

    for student in students:
        if c_name == student['이름']:
            print(student)
            f = True
            break
    if not f:
        print("없음")
except FileNotFoundError:
    print("학생 정보 파일이 없습니다.")
```

```python
def search_student():
    name_to_search = input("검색할 학생 이름: ")
    
    try:
        with open("students.json", "r", encoding="utf-8") as f:
            students = json.load(f)
            found = False
            
            for student in students:
                if student["이름"] == name_to_search:
                    print("학생 정보:", student)
                    found = True
            
            if not found:
                print("해당 학생을 찾을 수 없습니다.")
    except FileNotFoundError:
        print("학생 정보 파일이 없습니다.")

search_student()
```

```python
# 미션 5: JSON 데이터를 CSV로 변환
# 목표: JSON 파일의 데이터를 CSV 파일로 변환하기.
# 단계:
# students.json 파일을 읽어 학생 정보를 가져옵니다.
# students.csv 파일을 생성하고, JSON 데이터를 CSV 형식으로 변환하여 저장합니다.
# CSV 파일의 내용을 출력합니다.
import json
import csv

def json_to_csv(jsonfilename,csvfilename):
    try:
        with open(jsonfilename, 'r', encoding = 'utf-8') as f:
            students = json.load(f)

            with open(csvfilename, 'w', encoding = 'utf-8', newline='') as f:
                wr = csv.writer(f)
                wr.writerow(students[0].keys())
                wr.writerows([student.values() for student in students])
            print("CSV 파일이 성공적으로 생성되었습니다.")
    except FileNotFoundError:
            print("학생 정보 파일이 없습니다.")

json_to_csv("students.json","students.csv")
```

```python
import json
import csv

def json_to_csv(jsonfile, csvfile):
    try:
        with open(jsonfile, "r", encoding="utf-8") as json_file:
            students = json.load(json_file)
            
            with open(csvfile, "w", encoding="utf-8", newline='') as f:
                writer = csv.writer(f)
                writer.writerow(students[0].keys())
                for row in students:
                    writer.writerow(row.values())
            print("CSV 파일이 성공적으로 생성되었습니다.")
    except FileNotFoundError:
        print("학생 정보 파일이 없습니다.")

json_to_csv("students.json","students.csv")
```
