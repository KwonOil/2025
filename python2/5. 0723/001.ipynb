```python
# 쇼핑카트에서 상품목록 출력
'''
1번 : 사과
2번 : 바나나
3번 : 딸기
'''
cart = ['사과','바나나','딸기']
shopping_list = {f"{k}번":v for k,v in enumerate(cart,1)}
for k,v in shopping_list.items():
    print(f"{k} : {v}")
```

```python
'''
민수 : 85점
지영 : 92점
철수 : 78점
'''
students = ['민수','지영','철수']
scores = [85, 92, 78]

class_score = {k:v for k,v in zip(students,scores)}
for k,v in class_score.items():
    print(f"{k} : {v}점")
```

```python
codes = ['M1','M2','M3']
students = ['민수','지영','철수']
scores = [85, 92, 78]

for c in zip(codes, students, scores):
    print(c)

for code, name, score in zip(codes, students, scores):
    print(f"[{code}] {name} : {score}점")
```

```python
'''
수학 : 90
영어 : 92
과학 : 95
'''
subjects = ['수학','영어','과학']
scores = [
    [87, 90, 85],
    [92, 88, 89],
    [76, 95, 91]
]

max_scores = [max(score) for score in scores]
max_subject = {s:m for s,m in zip(subjects, max_scores)}
for k,v in max_subject.items():
    print(f"{k} 최고점 : {v}점")
```

```python
receipts = [12000,5300,21200,4300]
print(receipts)

# 컴프리헨션으로 리스트를 생성한 후 출력
even = [r for i, r in enumerate(receipts,1) if i % 2 == 0]
print(even)

result1 = sum(even)
print(result1)

# 제너레이터 표현식을 이용해 결과값만 생성 후 출력
result2 = sum(r for i, r in enumerate(receipts,1) if i % 2 == 0)
print(result2)

result3= (r for i, r in enumerate(receipts,1) if i % 2 == 0)
print(sum(result3))
```

```python
# 리스트 컴프리헨션
s1 = [x**2 for i, x in enumerate(range(10),1) if i % 2 == 0]
print(s1)

# 제너레이터 표현식(1회용)
s2 = (x**2 for i, x in enumerate(range(10),1) if i % 2 == 0)
print(list(s2))

print(s1)
print(list(s2))
```

```python
morning = ['민호','영수','지혜']
evening = ['수진','영희','준호']
result = [(a,b) for a,b in zip(morning,evening)]
print(result)
# 전치 : 행과 열을 바꾸는 것
f,s = zip(*result) # * : 언패킹
# f,s = zip(result[0], result[1], result[2])
r = list(zip(list(s),list(f)))
print(r)
```

```python
answers = [3, 1, 2, 4, 2, 3, 3, 1, 5]
student = [3, 1, 2, 1, 2, 1, 2, 1, 5]

corr = [1 if a == s else 0 for a, s in zip(answers, student)]
score = score.count(1) / len(answers) * 100

print(corr)
print(f"{score:.2f}점")
```

```python
customer_tuples = [
    ("C001", "김철수", 35, "남성",150000,"2024-03-15"),
    ("C002", "이영희", 35, "여성",280000,"2024-03-14"),
    ("C003", "박민수", 35, "남성",95000,"2024-03-13")
]
# 튜플데이터 : (고객ID, 이름, 나이, 성별, 구매금액, 구매날짜)
```

```python
keys1 = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict1 = {key : list(customer) for key, customer in zip(keys1, zip(*customer_tuples))}
dict1
```

```python
keys2 = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict2 = [dict(zip(keys2,customer)) for customer in customer_tuples]
dict2
```

```python
keys_id,*keys_info = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict3 = {id : dict(zip(keys_info,customer)) for id, *customer in customer_tuples}
dict3
```
