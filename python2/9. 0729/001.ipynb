```python
# 점수에 따라 학점 매핑후 학점 추가하고 json으로 변환
students = [
    {"name" : "홍길동", "score" : 95},
    {"name" : "김영희", "score" : 82},
    {"name" : "박철수", "score" : 67}
]

for student in students:
    if student['score'] >= 90:
        student['grade'] = 'A'
    elif student['score'] >= 80:
        student['grade'] = 'B'
    elif student['score'] >= 70:
        student['grade'] = 'C'
    elif student['score'] >= 60:
        student['grade'] = 'D'
    else:
        student['grade'] = 'F'
    

import json

json_students = json.dumps(students, ensure_ascii=False, indent=2)
print(json_students)
```

```python
# 10000원이 넘는 경우에만 10% 할인해서 최종가격 항목 추가
products = [
    {"id" : 1, "price" : 12000},
    {"id" : 2, "price" : 5500}
]
for product in products:
    price = product.get('price')
    if price >= 10000:
        product['final_price'] = round(price * 0.9)
    else:
        product['final_price'] = price

json_products = json.dumps(products, ensure_ascii=False, indent=2)
print(json_products)
```

```python
# 총점과 평균 항목 추가
students = [
    {"name" : "철수", "scores" : {"국어" : 85, "수학" : 90, "영어" : 80}},
    {"name" : "영희", "scores" : {"국어" : 95, "수학" : 85}}
]

for student in students:
    scores = student['scores']
    total = sum(scores.values())
    cnt = len(scores.values())
    avg = total / cnt if cnt > 0 else 0
    student['total'] = total
    student['avarage'] = avg

json_students = json.dumps(students, ensure_ascii=False, indent=2)
print(json_students)
```

```python
my_dict = {
    "10/26" : [100, 130, 50, 90],
    "10/27" : [10, 12, 10, 11]
}

def print_val(d, key, val):
    return [i for i in d[key] if i >= val]

print(print_val(my_dict, "10/26", 100))
print(print_val(my_dict, "10/27", 11))
```

```python
numbers = list(range(1,11))
# filter(함수, 리스트) : 리스트의 각 원소에 함수를 적용하여 True인 것만 추출하여 새로운 리스트를 만듦
r = list(map(lambda x : x % 2 == 0, numbers))
print(r)
v = list(filter(lambda x : x % 2 == 0, numbers))
print(v)
```

```python
def average(*args):
    return round(sum(args) / len(args)) if args else 0
scores = [95, 20, 30, 40, 55, 70, 98, 100, 50, 80, 70, 90, 0, 40, 60, 80]
print(average(*scores))
```

```python
def print_user_info(name, age, *args):
    print(f"이름 : {name}")
    print(f"나이 : {age}")
    if args:
        print("나머지 정보")
        for info in args:
            print(f"- {info}")
print_user_info("Alice", 30, "New York", "555-1234", "Likes Python")
print()
print_user_info("Bob", 25)
print()
```

```python
# 딕셔너리 가변인자
def print_user_info(**kwargs):
    print(type(kwargs))
    print(kwargs)
    for k, v in kwargs.items():
        print(f"{k} : {v}")

print_user_info(name = 'Alice', age = 30, city = 'NewYork')
```

```python
def greet(**kwargs):
    name = kwargs.get('name','Guest')
    print(f"Hello, {name}!")

greet(name = 'Bob')
greet()
```

```python
def custom_error(message, **kwargs):
    error_details = f"Error: {message}"
    for key, value in kwargs.items():
        error_details += f"\n{key}: {value}"
    return error_details

error_message = custom_error("File not found", filename="example.txt", line=42)
print(error_message)
```

```python
calculator = {
    'add' : lambda x, y : x + y,
    'sub' : lambda x, y : x - y,
    'mul' : lambda x, y : x * y,
    'div' : lambda x, y : x / y,
}
a = calculator['add'](10,5)
s = calculator['sub'](10,5)
m = calculator['mul'](10,5)
d = calculator['div'](10,5)
calculator['sum'] = lambda *args : sum(args)
sum1 = calculator['sum'](1,2,3,4,5,6,7,8,9)

print(f"add : {a}", f"sub : {s}", f"mul : {m}", f"div : {d}", f"sum : {sum1}", sep = '\n')
```
