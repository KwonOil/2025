```python
my_dict = {
    "이름" : "홍길동",
    "나이" : 25,
}
# 딕셔너리 값 가져오기
print(my_dict["이름"])
print(my_dict["나이"])
print(my_dict.get("성별", "성별 정보 없음"))

# 딕셔너리 값 추가
my_dict["성별"] = '여성'
print(my_dict)

# 딕셔너리 값 삭제
del my_dict['성별']
print(my_dict)

my_dict["성별"] = '남성'
print(my_dict)

if "성별" in my_dict:
    print(f"값 : {my_dict['성별']}")
else:
    print("key가 존재하지 않습니다")
```

```python
for key, value in my_dict.items():
    print(key, value)

print()

for key in my_dict.keys():
    print(key)

print()

for value in my_dict.values():
    print(value)

print()

print(list(my_dict.keys()))
print(list(my_dict.values()))
print(list(my_dict.items()))

print(type(my_dict.keys()))
print(type(list(my_dict.keys())))
```

```python
import matplotlib.pyplot as plt

monthly_avg_temp = {
    'January': -1,
    'February': 0,
    'March': 5,
    'April': 10,
    'May': 15,
    'June': 20,
    'July': 25,
    'August': 24,
    'September': 19,
    'October': 12,
    'November': 5,
    'December': 0
}

months = list(monthly_avg_temp.keys())
temps = list(monthly_avg_temp.values())

print(months, temps, sep="\n")

plt.figure(figsize=(10, 5))
plt.plot(months, temps, marker='o')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import requests
import json
apiurl = 'https://api.openweathermap.org/data/2.5/weather?q='
city = ['Seoul', 'Busan', 'Incheon', 'Daegu', 'Daejeon', 'Gwangju', 'Ulsan']
apikey = 'da6dae7bf19dd1d89fe7cad070e71b06'
    
def weather(num):
    urls = apiurl + city[num-1] + "&lang=kr&appid=" + apikey
    response  = requests.get(urls)
    data = json.loads(response.text)
    print(data)
    cityname = data['name']
    weather = data['weather'][0]['main']
    temp = int(data['main']['temp'] - 273.15)
    print(cityname+':', weather+',', str(temp)+'˚C')

print("1: Seoul / 2: Busan / 3: Incheon / 4: Daegu / 5: Daejeon / 6: Gwangju / 7: Ulsan")

inum = int(input("Choose city number: "))
weather(inum)
```

```python
import matplotlib.pyplot as plt
import json

apiurl = 'https://api.openweathermap.org/data/2.5/weather?q='
city = ['Seoul', 'Busan', 'Incheon', 'Daegu', 'Daejeon', 'Gwangju', 'Ulsan']
apikey = 'da6dae7bf19dd1d89fe7cad070e71b06'

temps = []

def weather2():
    for c in city:
        urls = apiurl + c + "&lang=kr&appid=" + apikey
        response  = requests.get(urls)
        data = json.loads(response.text)
        temps.append(round(data['main']['temp'] - 273.15, 2))

weather2()

print(city)
print(temps)

plt.figure(figsize=(10, 5))  # 플롯 크기 설정
plt.plot(city, temps, marker='o')  # 선 그래프 생성
plt.title('Weather')  # 제목 설정
plt.xlabel('City')  # x축 레이블 설정
plt.ylabel('temp(°C)')  # y축 레이블 설정
plt.grid()  # 그리드 추가
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()  # 플롯 표시
```

```python
import matplotlib.pyplot as plt
import json

apiurl = 'https://api.openweathermap.org/data/2.5/weather?q='
cities = ['Seoul', 'Busan', 'Incheon', 'Daegu', 'Daejeon', 'Gwangju', 'Ulsan']
apikey = 'da6dae7bf19dd1d89fe7cad070e71b06'

def get_weather_data():
    city_temperatures = []

    for city in cities:
        url = f"{apiurl}{city}&lang=kr&appid={apikey}"
        response = requests.get(url)
        data = json.loads(response.text)
        temp = int(data['main']['temp'] - 273.15)
        city_temperatures.append(temp)

    return city_temperatures

def plot_temperatures(city_temperatures):
    plt.figure(figsize=(10, 6))
    plt.bar(cities, city_temperatures, color='skyblue')
    plt.xlabel('Temperature (˚C)')
    plt.title('Temperatures in South Korean Cities')
    plt.grid(axis='x')
    plt.show()

city_temperatures = get_weather_data()
plot_temperatures(city_temperatures)
```
