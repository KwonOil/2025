```python
import urllib.request
from urllib.error import HTTPError, URLError

url = 'http://example.com'

try:
    response = urllib.request.urlopen(url)
    html = response.read().decode('utf-8')  # 응답을 문자열로 변환
    print("웹 페이지 내용:\n", html)
except HTTPError as e: # HTTP 오류
    print('HTTP 오류:', e.code)
except URLError as e:  # URL 오류
    print('URL 오류:', e.reason)
```

```python
import urllib.request
from urllib.parse import urlparse, urlencode, parse_qs, urlunparse

url = 'http://www.example.com/path?query=1#fragment'

# URL 파싱
parsed_url = urlparse(url)
print("파싱된 URL 구성 요소:")
print("프로토콜:", parsed_url.scheme)
print("호스트:", parsed_url.hostname)
print("경로:", parsed_url.path)
print("쿼리:", parsed_url.query)
print("프래그먼트:", parsed_url.fragment)

# 쿼리 문자열 인코딩(딕셔너리를 쿼리 문자열로 변환)
query_params = {'key1': 'value1', 'key2': 'value2'}
encoded_query = urlencode(query_params)
print("인코딩된 쿼리 문자열:", encoded_query)

# 새로운 URL 생성
# new_url = f"{parsed_url.scheme}://{parsed_url.hostname}{parsed_url.path}?{encoded_query}"
new_url = parsed_url._replace(query = encoded_query)
print("새로운 URL :", new_url)

# 언파싱
unparsed_url = urlunparse(new_url)
print("새로운 URL :", unparsed_url)

# URL 쿼리 문자열 파싱
query_dict = parse_qs(parsed_url.query)
print(query_dict)

query_dict_single = {key: value[0] for key, value in query_dict.items()}
print(query_dict_single)
```

```python
## post로 데이터 전송
import urllib.request
import json

url = "http://httpbin.org/post"
data = {
    'name': 'John Doe',
    'age': 30
}

# encode: 문자열을 바이트 문자열(바이너리)로 변환
# decode: 바이트 문자열(바이너리)을 문자열로 변환
data_encoded = json.dumps(data).encode('utf-8')
req = urllib.request.Request(url, data=data_encoded, method='POST')
req.add_header('Content-Type', 'application/json')

with urllib.request.urlopen(req) as response:
    result = response.read().decode('utf-8')
    print(result)
```

```python
## post로 데이터 전송
import urllib.request
import json

# 테스트 서버 (데이터를 받을 수 있는 서버)
# url = "http://httpbin.org/post"
url = "http://localhost:5000/post"

data = {
    'name': '권오일',
    'age': 33,
    'city': '경기도',
    'hobbies': ['독서', '영화 감상', '게임'],
    'gender': '남성'
}

# encode: 문자열을 바이트 문자열(바이너리)로 변환
# decode: 바이트 문자열(바이너리)을 문자열로 변환
data_encoded = json.dumps(data).encode('utf-8')

# 서버에 데이터 전송 후 req로 응답 받기
req = urllib.request.Request(url, data=data_encoded, method='POST')
req.add_header('Content-Type', 'application/json')


with urllib.request.urlopen(req) as response:
    result = response.read().decode('utf-8')

# result를 JSON으로 파싱
mydata = json.loads(result)
mydata
```
