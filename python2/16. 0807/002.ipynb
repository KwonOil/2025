```python
# 백업
# 1) 백업폴더가 존재하는지
# 2) 백업폴더가 없으면 생성 아니면 계속 진행
# 3) 원본폴더의 파일목록 불러오기
# 4) 파일목록 순서대로 돌기
# 5) 전체경로 만들기(파일)
# 6) 목록이 파일인지 확인 (파일일 경우에만 백업진행)
# 7) 작업폴더의 파일을 백업폴더의 파일에 저장
import os
import shutil
source_directory = os.getcwd()
backup_directory = 'Z:/html/sample_backup'

def backup_files(src,dst):
    # 1)
    if not os.path.exists(dst):
        # 2)
        os.makedirs(dst, exist_ok = True)
    
    # 4)
    for filename in os.listdir(src):
        # 3)
        # print(filename)
        # 5)
        source_path = os.path.join(src,filename)
        backup_path = os.path.join(dst,filename)

        # 6)
        if os.path.isfile(source_path):
            # 7)
            shutil.copy(source_path, backup_path)
            if os.path.exists(backup_path):
                print(f"{backup_path} - 백업 완료")
            else:
                print(f"{backup_path} - 백업 실패")

backup_files(source_directory, backup_directory)
```

```python
# 파일 분류 (확장자별 폴더로 파일 정리[이동])
# 1) 타겟 경로내 파일목록 불러오기  
# 2) 파일인지 아닌지 확인 (파일인 경우만 진행)  
# 3) 확장자 확인(분리)  
# 4) 타겟 경로와 확장자 합치기 (생성할 확장자 폴더 경로)  
# 5) 확장자 폴더가 존재하지 않을때만 확장자 폴더생성  
# 6) 타겟 경로와 경로내 파일 연결 (절대경로 생성)  
# 7) 해당 확장자 경로로 파일 이동
def organize_files(target_dir):
    for filename in os.listdir(target_dir):
        target_file = os.path.join(target_dir, filename)
        if os.path.isfile(target_file):
            ext = filename.split('.')[-1]
            ext_dir = os.path.join(target_dir, ext)
            if not os.path.exists(ext_dir):
                os.makedirs(ext_dir)
                print(f"확장자 폴더 생성 - {ext}")
            new_file = os.path.join(ext_dir, filename)
            shutil.move(target_file, new_file)
            print(f"{filename} 파일이 {ext_dir}로 이동되었습니다")
target_dir = os.getcwd() + '/example_dir3'
organize_files(target_dir)
```

```python
import shutil
import os

# 압축할 디렉토리 생성
os.makedirs("my_folder/subfolder", exist_ok=True)

# 임의의 파일 생성
with open("my_folder/subfolder/file1.txt", "w") as f:
    f.write("This is file 1.")

with open("my_folder/subfolder/file2.txt", "w") as f:
    f.write("This is file 2.")

# 아카이브 생성
shutil.make_archive("my_archive", 'zip', "my_folder")

print("아카이브가 생성되었습니다: my_archive.zip")
```
