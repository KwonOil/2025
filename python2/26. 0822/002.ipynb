```python
# 1) selenium 기본 라이브러리
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# 2) Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--headless') # 창 열리지 않게
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# 3) Chrome 드라이버 설정
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

#############################################################

## 동적(추후 자바스크립트가 내용을 로딩) 사이트 파싱
url = "https://www.tistory.com/?category=living"
driver.get(url)

time.sleep(5)

items = driver.find_elements(By.CSS_SELECTOR,'div.item_group')

posts = []

for item in items:
    title = item.find_element(By.CSS_SELECTOR, '.tit_cont').text.strip()
    desc = item.find_element(By.CSS_SELECTOR, '.wrap_desc').text.strip()
    author = item.find_element(By.CSS_SELECTOR, '.link_blog .txt_g').text.strip()

    posts.append({
        'title': title,
        'description': desc,
        'author': author
    })

driver.quit()
#posts

html_content = f'''
<html>
<head>
    <style>
        .blog-container {{
            max-width: 900px;
            margin: 20px auto;
            font-family: 'Noto Sans KR', sans-serif;
        }}
        .blog-post {{
            background: #fff;
            padding: 20px;
            margin: 15px 0;
            box-shadow: 0 2px 5px rgba(0,0,0,0.1);
            border-radius: 8px;
        }}
        .blog-title {{
            font-size: 20px;
            color: #333;
            margin-bottom: 10px;
        }}
        .blog-desc {{
            color: #666;
            line-height: 1.6;
        }}
        .blog-author {{
            color: #888;
            font-size: 14px;
            margin-top: 10px;
        }}
    </style>
</head>
<body>
    <div class="blog-container">
'''

for post in posts:
    html_content += f'''
        <div class="blog-post">
            <div class="blog-title">{post['title']}</div>
            <div class="blog-desc">{post['description']}</div>
            <div class="blog-author">작성자: {post['author']}</div>
        </div>
    '''

html_content += '''
    </div>
</body>
</html>
'''

# HTML 파일 저장
with open('tistory_posts2.html', 'w', encoding='utf-8') as f:
    f.write(html_content)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
from datetime import datetime
from bs4 import BeautifulSoup
import json

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36')

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)


keyword = "빅데이터"
url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"

driver.get(url)

WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, 'li.bx .view_wrap'))
)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# 페이지 소스 가져오기 (page_source)
html = driver.page_source
driver.quit()

soup = BeautifulSoup(html, 'html.parser')

posts = []

items = soup.select('li.bx .view_wrap')

for item in items:
    title = item.select_one('.title_area a').text
    link = item.select_one('.title_area a')['href']
    summary = item.select_one('.dsc_area').text

    posts.append({
        'title': title,
        'link': link, 
        'summary': summary
    })

#posts

with open(f'naver_blog_{keyword}_{datetime.now().strftime("%Y%m%d%H%M%S")}.json', 'w', encoding='utf-8') as f:
    json.dump(posts, f, ensure_ascii=False, indent=2)
```
