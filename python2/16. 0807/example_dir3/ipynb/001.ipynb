문제 : 식료품 구매 영수증 계산기

우유 : 2500
빵 : 3000
계란 : 5000
과일 : 8000

식료품 구매 영수증을 계산하기 위한 프로그램을 작성합니다.
사용자가 구매한 품목과 가격을 입력받아 소계, 세금, 총액을 계산하는 기능을 구현

요구사항

calculate_receipt 함수 : 
 - 품목과 가격 정보를 활용해 소계, 10% 세금, 총액을 계산합니다.
 - 소계는 모든 품목의 가격 합계입니다.
 - 세금은 소계의 10%입니다.
 - 총액은 소계와 세금을 합한 값입니다.

출력 형식 : 
 - 소계 : 원 단위로 첨 단위 구분기호를 사용하여 출력합니다.
 - 부가세 : 소수점 없이 출력합니다.
 - 총액 : 소수점 없이 출력합니다.
<결과>
소계 : 18,500원
부가세 : 1,850원
총액 : 20,350원
</pre>

```python
products = {
    '우유' : 2500,
    '빵' : 3000,
    '계란' : 5000,
    '과일' : 8000
}

def calculate_receipt(products):
    total = sum(products.values())
    tax_rate = 0.1
    tax = total * tax_rate
    result = total + tax
    return total, tax, result

total, tax, result = calculate_receipt(products)
print(f"소계 : {total:,.0f}원\n세금 : {tax:,.0f}원\n총액 : {result:,.0f}원")
```

문제 : BMI 계산기
BMI(체질량지수)는 체중과 신장을 기반으로 개인의 체중 상태를 평가하는 지표입니다.
이 프로그램은 사용자의 체중과 신장을 입력받아 BMI를 계산하고, 그에 따른 체중 상태를 출력합니다.

저체중 : BMI < 18.5
정상 : 18.5 <= BMI < 25
과체중 : 25 <= BMI < 30
비만 : 30 <= BMI

공식 : bmi = 체중 / ((신장/100)**2)

```python
weight = int(input("몸무게를 입력하세요"))
height = int(input("키를 입력하세요"))

def bmi(w,h):
    return w / ((h / 100)**2)

def status_bmi(bmi):
    if bmi < 18.5: return '저체중'
    elif bmi < 25: return '정상'
    elif bmi < 30: return '과체중'
    else: return '비만'

print(f"몸무게 : {weight}kg, 키 : {height}cm")
print(f"BMI : {bmi(weight,height):.2f}")
print("상태 : " + status_bmi(bmi(weight,height)))
```

문제 : 월별 전기 사용량 비용 계산기

가정의 월별 전기 사용량에 따라 전기 요금을 계산하는 프로그램을 작성합니다.
전기 사용량에 따라 요금 단가가 달라지며, 기본 요금이 추가됩니다.

전기 사용량(kWh)을 입력받아 요금을 계산합니다.
요금 계산 기준은 다음과 같습니다:
- 사용량 200 kWh 이하 : 단가 120원 / 기본요금 : 910
- 사용량 201 kWh ~ 400 kWh : 단가 214.6원 / 기본요금 : 1,600원
- 사용량 401 kWh ~ : 단가 307.3원 / 기본요금 : 7,300원

```python
using = int(input("전기 사용량을 입력하세요"))

def cal_using(using):
    if using <= 200:
        return 120 + using * 910
    elif using <= 400:
        return 214.6 + using * 1600
    else:
        return 307.3 + using * 7300

print(f"{cal_using(using):,}원")
```
