```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_train = pd.read_csv('train.txt', sep='\t', encoding='utf-8', engine='python')
df_train.info()
df_train.head()
```

```python
df_test = pd.read_csv('ratings_test.txt', sep='\t', encoding='utf-8', engine='python')
df_test.head()
```

```python
df_train = df_train[df_train['document'].notnull()]
df_test = df_test[df_test['document'].notnull()]
df_train.info()
```

```python
df_train.head()
```

```python
df_train['label'].value_counts()
```

```python
import re
df_train['document'] = df_train['document'].apply(lambda x:re.sub(r'[^ ㄱ-ㅣ가-힣]+', ' ', x))
```

```python
df_train.info()
```

```python
df_test['label'].value_counts()
```

```python
df_test['document'] = df_test['document'].apply(lambda x:re.sub(r'[^ ㄱ-ㅣ가-힣]+', ' ', x))
```

```python
df_test = df_test[df_test['document'].str.strip() != ""]
df_test
```

```python
from konlpy.tag import Okt
okt = Okt()
okt.morphs("나 거주 경기도.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def okt_tokenizer(text):
    tokens = okt.morphs(text)
    return tokens

tfidf = TfidfVectorizer(tokenizer=okt_tokenizer, ngram_range=(1, 2), min_df=3, max_df=0.9)
tfidf.fit(df_train['document'])
nsmc_train_tfidf = tfidf.transform(df_train['document'])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state = 0)
model.fit(nsmc_train_tfidf, df_train['label'])
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C': [1, 3, 3.5, 4, 4.5, 5]}
SA_lr_grid_cv = GridSearchCV(model, param_grid=params, cv=3, scoring='accuracy', verbose=1)
SA_lr_grid_cv.fit(nsmc_train_tfidf, df_train['label'])
```

```python
print(SA_lr_grid_cv.best_params_, round(SA_lr_grid_cv.best_score_, 4))
```

```python
SA_lr_best = SA_lr_grid_cv.best_estimator_
SA_lr_best
```

### 모델 평가

```python
nsmc_test_tfidf = tfidf.transform(df_test['document'])
```

```python
test_predict = SA_lr_best.predict(nsmc_test_tfidf)
test_predict
```

```python
from sklearn.metrics import accuracy_score
print("감성 분석 정확도 : ",round(accuracy_score(df_test['label'], test_predict),3))
```

### 감성 예측

```python
st = input('문장입력 >> ')
st2 = re.compile(r'[ㄱ-ㅣ가-힣]+').findall(st)
st3 = [' '.join(st2)]

st_tfidf = tfidf.transform(st3)

st_predict = SA_lr_best.predict(st_tfidf)[0]

if(st_predict == 0):
    print(st, '==> 부정 감성')
else:
    print(st, '==> 긍정 감성')
```
