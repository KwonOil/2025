```python
store_sales_dict = {
    "S001": {"지역": "서울", "일일판매액": 5000000, "방문객수": 200, "평균구매액": 25000, "직원수": 8},
    "S002": {"지역": "부산", "일일판매액": 3500000, "방문객수": 150, "평균구매액": 23000, "직원수": 6},
    "S003": {"지역": "대구", "일일판매액": 2800000, "방문객수": 120, "평균구매액": 23500, "직원수": 5},
    "S004": {"지역": "서울", "일일판매액": 6000000, "방문객수": 300, "평균구매액": 20000, "직원수": 10},
}
# 지역별 직원 1인당 매출액을 구하시오

result = {}

for info in store_sales_dict.values():
    if info["지역"] not in result:
        result[info["지역"]] = []
    result[info["지역"]].append(info["일일판매액"] / info["직원수"])

add_result = {k : f"{sum(v):,.0f}" for k, v in result.items()}
print(add_result)
add_result2 = {k : round(sum(v)) for k, v in result.items()}
print(add_result2)
```

```python
feedback_dict = {
    "R001": {"제품ID": "P001", "평점": 4.5, "리뷰내용": "매우 만족합니다", "작성일": "2024-03-15", "추천수": 10},
    "R002": {"제품ID": "P001", "평점": 3.5, "리뷰내용": "보통입니다", "작성일": "2024-03-14", "추천수": 5},
    "R003": {"제품ID": "P002", "평점": 5.0, "리뷰내용": "최고의 제품", "작성일": "2024-03-13", "추천수": 15}
}
# 제품별 평균 평점 계산
result4 = {}
for info in feedback_dict.values():
    if info['제품ID'] not in result4:
        result4[info['제품ID']] = {"평점합계" : 0, "리뷰수" : 0, "평균평점" : 0}
    result4[info['제품ID']]["평점합계"] += info["평점"]
    result4[info['제품ID']]["리뷰수"] += 1

for pid, rat in result4.items():
    rat["평균평점"] = rat["평점합계"] / rat["리뷰수"]

result4
```
