```python
class Human:
    
    # 생성자: 인스턴스가 생성될 때 자동 호출됨
    def __init__(self, name, age, gender):
        self.__name = name
        self.__age = age
        self.__gender = gender
        print(f"{self.__name}님 환영합니다.")

    # 소멸자: 인스턴스가 삭제될 때 자동 호출됨
    def __del__(self):
        print(f"정말 지우시나요???")

    def showInfo(self):
        print(f"이름: {self.__name}, 나이: {self.__age}, 성별: {self.__gender}")
    
    def setInfo(self, name, age, gender):
        self.__name = name
        self.__age = age
        self.__gender = gender

    def getInfo(self):
        return self.__name, self.__age, self.__gender

a = Human("홍길동", 31, "남자")
a.showInfo()  # 정보 출력
a.setInfo("김영희", 25, "여자")  # 정보 수정
print(a.getInfo())  # 수정된 정보 출력
a.showInfo()
del a  # 인스턴스 삭제
```

```python
### 장바구니 리스트

'''
1: 항목 추가
2: 항목 삭제
3: 목록 보기
4: 종료
원하는 작업을 선택하세요:
'''
## 항목 : 오이, 호박, 두부, 감자, 배추 ...

basket = []
def show_menu():
    print("=== 메뉴 ===")
    print("1: 항목 추가")
    print("2: 항목 삭제")
    print("3: 목록 보기")
    print("4: 종료")

def is_exist(name):
    if not basket:
        return False
    else:
        for product in basket:
            if name not in product['name']:
                return False
            else:
                return True

def add_product():
    name = input("추가할 상품 이름을 입력하세요 : ")

    if is_exist(name) == False:
        price = int(input("추가할 상품 가격을 입력하세요 : "))
        basket.append({ 'name' : name, 'price' : price, 'amount' : 1 })
    else:
        for product in basket:
            product['amount'] += 1

def del_product():
    name = input("삭제할 상품 이름을 입력하세요 : ")

def show_products():
    for product in basket:
        print(f"이름 : {product['name']}, 가격 : {product['price']}, 수량 : {product['amount']}, 수량type : {type(product['amount'])}")
while(True):
    show_menu()

    choice = input("원하는 작업을 선택하세요")
    
    if choice == '1':
        add_product()
    elif choice == '2':
        pass
    elif choice == '3':
        show_products()
    elif choice == '4':
        break
```
