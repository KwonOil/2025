```python
import random
num = random.randint(1, 10)
life = 3

while True:
    usernum = input("1~10 사이의 숫자를 입력하세요: ")

    if usernum == "":
        print("게임종료!")
        break

    if int(usernum) == num:
        print(f"정답입니다. (답: {num})")
        break
    else:
        life -= 1
        if life == 0:
            print(f"실패했습니다. (답: {num})")
            break
        elif int(usernum) > num:
            print("입력한 숫자가 랜덤 숫자보다 큽니다.")
        elif int(usernum) < num:
            print("입력한 숫자가 랜덤 숫자보다 작습니다.")

    print("남은 기회: ", life)
```

### math

```python
import math

# 1. ceil() - 올림
print(math.ceil(1.4))    # 2
print(math.ceil(1.7))    # 2
print(math.ceil(-1.4))   # -1

# 2. round() - 반올림
print(round(1.4))        # 1
print(round(1.7))        # 2
print(round(1.5))        # 2
print(round(-1.5))       # -2

# 3. floor() - 내림
print(math.floor(1.7))   # 1
print(math.floor(1.4))   # 1
print(math.floor(-1.7))  # -2

# 4. sqrt() - 제곱근
print(math.sqrt(4))      # 2.0
print(math.sqrt(2))      # 1.4142135623730951

# 5. factorial() - 계승
print(math.factorial(5)) # 120 (5! = 5 × 4 × 3 × 2 × 1)
print(math.factorial(3)) # 6 (3! = 3 × 2 × 1)

# 6. pi - 원주율
print(math.pi)           # 3.141592653589793
```

[strftime과 strptime]

strftime과 strptime은 Python의 datetime 모듈에서 날짜와 시간을 문자열로 변환하거나
문자열을 날짜 및 시간 객체로 변환하는 데 사용되는 메서드입니다.

1. strftime (String Format Time)
strftime은 datetime 객체를 문자열로 변환할 때 사용됩니다.
날짜와 시간을 특정 형식으로 출력할 수 있도록 포맷을 지정할 수 있습니다.

(사용법)

datetime_object.strftime(format)

주요 포맷 코드
%Y: 4자리 연도 (예: 2023)
%y: 2자리 연도 (예: 23)
%m: 2자리 월 (01-12)
%d: 2자리 일 (01-31)
%H: 24시간제 시 (00-23)
%I: 12시간제 시 (01-12)
%M: 분 (00-59)
%S: 초 (00-59)
%A: 요일 이름 (예: Monday)
%B: 월 이름 (예: January)

2. strptime (String Parse Time)
strptime은 문자열을 날짜 및 시간 객체로 변환할 때 사용됩니다.
입력 문자열과 일치하는 포맷을 지정해야 합니다.
</pre>

### timedelta
- 두 날짜 또는 시간 간의 차이를 표현하는 데 사용

```python
from datetime import datetime, timedelta
now = datetime.now()
print("현재 시간 : ", now)
# 27일 후
fdate = now + timedelta(days = 27)
print("27일 후 :", fdate)
# 5시간 전
past_time = now - timedelta(hours = 5)
print("5시간 전 : ", past_time)
```

```python
input_date = input("날짜를 입력하세요 (YYYY-MM-DD 형식)")
date_obj = dt.datetime.strptime(input_date,"%Y-%m-%d")
# 10일 후
next_time = date_obj + timedelta(days=10)
next_time = next_time.strftime("%Y-%m-%d")
# 5일 전
pre_time = date_obj - timedelta(days=5)
pre_time = pre_time.strftime("%Y-%m-%d")
print("10일 후 :", next_time)
print("5일 전 :", pre_time)
```

```python
input_date = input("날짜를 입력하세요 (YYYY-MM-DD 형식)")
date_obj = dt.datetime.strptime(input_date,"%Y-%m-%d")
print("현재 :", date_obj.strftime("%Y-%m-%d %H:%M:%S"))

# 10일 5시간 20분 후
next_time = date_obj + timedelta(days = 10, hours = 5, minutes = 20)

# 5일 2시간 15분 30초 전
pre_time = date_obj - timedelta(days = 5, hours = 2, minutes = 15, seconds = 30)

print("10일 후 :", next_time)
print("5일 전 :", pre_time)
```
