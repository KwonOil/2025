```python
n = 10

for i in range(1,n):
    k = i if i <= n // 2 else n - i
    txt = '*' * (2 * k - 1)
    print(f"|{txt:^{n-1}}|")
    # print(f"i:{i}\tk:{k}\tn-i:{n-i}")
```

```python
length = 20
half = length//2

for i in range(1,length):
    k = i if i <= half else length - i
    bnk = ' ' * (half - i if i <= half else half - k)
    txt = '*' * (2 * k - 1)
    print(bnk + txt + bnk)
```

```python
# 1부터 100까지의 합
total = 0
limit = 10000
for i in range(limit+1):
    total += i
print(f"1부터 {limit:,}까지의 총합 : {total:,}!")
```

```python
# 1부터 100까지의 짝수의 합
total = 0
limit = 1000
for i in range(0,limit+1,2):
    total += i
print(f"1부터 {limit}까지의 총합 : {total}!")
```

```python
# 1부터 100까지 4의 배수의 갯수는?
total = 0
limit = 100
cnt = 0
for i in range(4,limit+1,4):
    if( i % 4 == 0):
        total += i
        cnt += 1
        print(i, end = " ")
print(f"\n4의 배수의 {limit}까지 총합 : {total}, 갯수는 {cnt}개")
```

```python
# ~번 학생 합격입니다
scores = [90, 25, 67, 45, 80]
for i in range(len(scores)):
    if(scores[i] >= 60):
        print(f"{i+1}번학생 합격입니다. [{scores[i]}점]")
    else:
        print(f"{i+1}번학생 불합격입니다. [{scores[i]}점]")
```

```python
# for i, s in enumerate(scores): # 인덱스를 0부터 시작(기본값)
for i, s in enumerate(scores, 1): # 인덱스를 1부터 시작
    if(s >= 60):
        print(f"{i}번학생 합격입니다. [{s}점]")
    else:
        print(f"{i}번학생 불합격입니다. [{s}점]")
```

```python
std = ['홍길동','이순신','강감찬','김구','세종대왕']
scores = [90, 25, 67, 45, 80]
for i, score in enumerate(scores):
    if score >= 60:
        print(f"{std[i]} 학생({score}점) 합격입니다")
    else:
        print(f"{std[i]} 학생({score}점) 불합격입니다")
```

```python
std = ['홍길동','이순신','강감찬','김구','세종대왕']
scores = [90, 25, 67, 45, 80]
for s, score in zip(std, scores):
    if score >= 60:
        print(f"{s} 학생({score}점) 합격입니다")
    else:
        print(f"{s} 학생({score}점) 불합격입니다")
```

```python
std = ['홍길동','이순신','강감찬','김구','세종대왕']
scores = [90, 25, 67, 45, 80]
for i, (s, score) in enumerate(zip(std, scores), 1):
    if score >= 60:
        print(f"{i}.{s} 학생({score}점) 합격입니다")
    else:
        print(f"{i}.{s} 학생({score}점) 불합격입니다")
```

```python

accounts = [8500, 12000, 15000, 7000]
rate = 0.01 # 이자 : 1%
for account in accounts:
    print(f"잔고 : {account:,}원", end = "\t")
    if account >= 10000: # 10000원 이상일 때만 이자 계산
        interest = account*rate
        print(f"이자 : {interest}원")
    else:
        print(f"이자 : 없음")
```
