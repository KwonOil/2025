```python
import tensorflow as tf

print("TF:", tf.__version__)
print("Physical GPUs:", tf.config.list_physical_devices('GPU'))

# ✅ 메모리 점유 폭주 방지 (가능하면 활성화)
gpus = tf.config.list_physical_devices('GPU')
for g in gpus:
    try:
        tf.config.experimental.set_memory_growth(g, True)
    except Exception as e:
        print("set_memory_growth error:", e)

# 더미 모델로 한 번 돌려보기
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),          # 예시 입력
    layers.Conv2D(16, 3, activation='relu'),
    layers.Flatten(),
    layers.Dense(10, activation='softmax'),
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

import numpy as np
x = np.random.rand(64, 28, 28, 1).astype('float32')
y = np.random.randint(0, 10, size=(64,))
history = model.fit(x, y, epochs=1, verbose=1)  # ✅ GPU가 보이면 자동 가속
```
