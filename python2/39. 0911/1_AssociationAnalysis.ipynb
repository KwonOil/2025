```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_csv('retail_dataset.csv')
df.info()
df
```

```python
items = set()
for col in df:
    print(df[col].unique())
    items.update(df[col].unique())
    print('-'*20)
items.remove(np.nan)
print(items)
```

```python
# 원-핫 인코딩 커스텀
itemset = set(items)

encoding = []
for index, row in df.iterrows():
    rowset = set(row)
    labels = {}
    # 구매하지 않은 상품 리스트
    dismatching = list(itemset - rowset)
    # 구매한 상품 리스트
    matching = list(itemset.intersection(rowset))

    for i in dismatching:
        labels[i] = 0
    for j in matching:
        labels[j] = 1

    encoding.append(labels)

result = pd.DataFrame(encoding)
result
```

```python
result.info()
```

```python
freq_items = apriori(result, min_support=0.2, use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules.head()
```

### with gpt

```python
row_lists = df.values.tolist()
products = [[x for x in row if not pd.isna(x)] for row in row_lists]
products
```

```python
# 원-핫 인코딩
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_array = te.fit(products).transform(products)
result = pd.DataFrame(te_array, columns=te.columns_)

# Apriori 알고리즘으로 빈발 항목 집합 찾기
frequent_itemsets = apriori(result, min_support=0.3, use_colnames=True)

# 연관 규칙 생성(gpt)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())

# 연관 규칙 생성(선생님)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules.head()
```
