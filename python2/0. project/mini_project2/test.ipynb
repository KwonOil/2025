### 데이터 불러오기

```python
import pandas as pd

df_seoul = pd.read_csv('csv/delay_proof_30min.csv')
df_korail = pd.read_csv('csv/korail_delay_30min.csv')
df2_seoul = df_seoul.copy()
df2_korail = df_korail.copy()
```

### 데이터 전처리

```python
df2_seoul.loc[df2_seoul['노선'] == '신정지선',['노선','방향']] = ['2호선','신정지선']
df2_seoul = df2_seoul.drop_duplicates(subset=['날짜', '노선', '방향'], keep='first').reset_index(drop=True)
# cond = (df2.iloc[:, 3:].isna().all(axis=1)) & (df2['방향'] == '외선')
# 조건을 만족하는 행만 제거
# df2 = df2[~cond]
# df2[df2['날짜'] == '2025-09-10']
# df2[(df2['노선'] == '2호선') & (df2['방향'] == '외선')]
df2_seoul
```

### 형변환

```python
for col_name in df2_seoul.columns[3:]:
    df2_seoul[col_name] = df2_seoul[col_name].str.split('분').str[0]
df2_seoul
```

```python
for col_name in df2_korail.columns[3:]:
    df2_korail[col_name] = df2_korail[col_name].str.split('분').str[0]
df2_korail
```

### 결측치 처리

```python
df2_seoul.fillna(0, inplace=True)
df2_seoul
```

```python
df2_korail.fillna(0, inplace=True)
df2_korail
```

### 형 변환2

```python
df2_seoul['날짜'] = pd.to_datetime(df2_seoul['날짜'].str.split('(').str[1].str.split(')').str[0], errors="coerce")
# df2['날짜'] = pd.to_datetime(df2['날짜'], errors="coerce")
df2_seoul.iloc[:, 1:3] = df2_seoul.iloc[:, 1:3].astype(str)
df2_seoul.iloc[:, 3:] = df2_seoul.iloc[:, 3:].astype(int)
df2_seoul
```

```python
df2_korail['날짜'] = pd.to_datetime(df2_korail['날짜'], errors="coerce").dt.date
df2_korail.iloc[:, 1:3] = df2_korail.iloc[:, 1:3].astype(str)
df2_korail.iloc[:, 3:] = df2_korail.iloc[:, 3:].astype(int)
df2_korail
```

### csv로 저장

```python
df2_seoul.to_csv('csv/delay_subway.csv', index=False)
df2_korail.to_csv('csv/delay_subway2.csv', index=False)
```

### 데이터 병합

```python
sql = """
DROP TABLE IF EXISTS delay_subway3;
CREATE TABLE delay_subway3 AS
SELECT * 
FROM delay_subway
WHERE 날짜 IN (SELECT 날짜 FROM delay_subway2)
UNION
SELECT * 
FROM delay_subway2
WHERE 날짜 IN (SELECT 날짜 FROM delay_subway)
ORDER BY 날짜 DESC, 노선;
"""
with engine.connect() as conn:
    # 여러 문장을 실행하려면 text() 안에 그대로 넣고 execute 가능
    for stmt in sql.strip().split(";"):
        if stmt.strip():  # 빈 문장 제외
            conn.execute(text(stmt))
    conn.commit()  # DDL 실행 후 커밋
```
