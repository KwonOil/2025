```python
# slice 예제
text = "동해물과 백두산이 마르고 닳도록 동서남북"
print("##" + text[5:9] + "##")
print("##" + text[5:9],"##",sep = "")
print("##" + text[:2] + "##")
print("##" + text[:-1] + "##")
print("##" + text[::2] + "##")
print("##" + text[1::2] + "##")
```

```python
print(len(text))
print(text.count("동"))
print(text[len(text)-1])
```

```python
mail = input("이메일 주소를 입력하세요:")
pos = mail.find("@")
id = mail[:pos]
domain = mail[pos+1:]
print("아이디 : ", id.strip())
print("도메인 : ", domain.strip())
```
