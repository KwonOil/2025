### argmax(), argmin() 메서드
* 배열에서 가장 큰 값 또는 가장 작은 값의 위치(인덱스)를 찾아준다.
  * argmax() : 가장 큰 값의 위치(인덱스) 찾기
  * argmin() : 가장 작은 값의 위치(인덱스) 찾기
* 기본적으로 배열을 평탄화(flatten)하여 인덱스를 반환

```python
import numpy as np
arr = np.array([[11, 12, 13, 14, 15],
                [21, 22, 23, 24, 25]])
print(arr.shape)

print(arr.max())
print(arr.min())

print(np.argmax(arr))
print(np.argmax(arr, axis = 0)) # 행
print(np.argmax(arr, axis = 1)) # 열
print('------------')
print(np.argmin(arr))
print(np.argmin(arr, axis = 0)) # 행
print(np.argmin(arr, axis = 1)) # 열
```

```python
import numpy as np
scores = np.array([88, 92, 79, 95, 85])

x = np.argmax(scores)
y = np.argmin(scores)
print(x, y)

x1 = scores.argmax()
y1 = scores.argmin()
print(x1, y1)
```

```python
import numpy as np

# 학생 성적: 행은 학생, 열은 과목
scores = np.array([[88, 92, 79],    # 학생1
                   [85, 90, 95],    # 학생2
                   [78, 88, 80]])   # 학생3

max_indices = np.argmax(scores, axis=0)
min_indices = np.argmin(scores, axis=0)
print("각 과목 최고 성적 학생 인덱스:", max_indices)
print("각 과목 최저 성적 학생 인덱스:", min_indices) 

scores.shape[1] # (3, 3)

# 최고 성적 값 추출
max_val = scores[[0,0,1], [0,1,2]]
print(max_val)

max_val2 = scores[max_indices, np.arange(scores.shape[1])]
print(max_val2)
```

```python
import numpy as np
x = np.array([18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13])

print(np.sort(x))             # 오름차순 정렬
print(np.sort(x)[::-1])       # 내림차순 정렬
print(len(x))                 # 데이터 개수
print(np.sum(x))              # 데이터 합계
print(np.mean(x))             # 평균
print(np.var(x))              # 분산 (데이터의 흩어짐 정도를 나타내며, 평균으로부터의 거리의 제곱을 평균한 값)
print(np.std(x))              # 표준편차 (분산의 제곱근으로, 원래 단위로 데이터의 변동성을 나타냄)
print(np.max(x))              # 최대값
print(np.min(x))              # 최소값
print(np.median(x))           # 중앙값 (데이터셋의 갯수를 기준으로 중앙에 위치한 값)
print(np.percentile(x, 25))   # 1사분위수
print(np.percentile(x, 50))   # 2사분위수
print(np.percentile(x, 75))   # 3사분위수
print(np.percentile(x, 100))  # 4사분위수
```

* loc
  * .loc[행] : 인덱스 레이블로 지정 (숫자 또는 문자, index 아님 주의)
    * df.loc[0]     # 시리즈 형태
    * df.loc[[0]]      # 데이터프레임 형태
    * df.loc[[0,1,3]]     # 여러 행을 한꺼번에 보려면 대괄호를 포함
    * df.loc[0:3] # 슬라이싱을 이용할 때는 대괄호 제거해야
      * df.loc[0:3] # 0 ~ 3번 레이블까지를 의미 --> 4개의 데이터를 보여준다 (0,1,2,3)
      * df.iloc[0:3] # 0 ~ 3 index까지를 의미 ---> 3개의 데이터를 보여준다 (0,1,2)
  * .loc[행, 열]
    * df.loc[0, '성명']     # 시리즈 형태
    * df.loc[[0], ['성명']]     # 데이터프레임 형태 (복수개 사용시에는 반드시 대괄호 필요)
    * df.loc[0:3, ['성명','급여']]  
    * df.loc[0:3, '성명':'급여']# 슬라이싱 사용시 대괄호 반드시 제거
  * .loc[조건, 열]
    * df.loc[df['급여']>=5000000] # '급여' 컬럼의 데이터에서 5백만원 이상인 행
    * df.loc[df['급여']>5000000, ['성명']] # '급여' 컬럼의 데이터에서 5백만원 이상인 행에 대해 '성명' 컬럼만
    * 파생 데이터 생성
      * df['체질량']=' '
      * df.loc[df['몸무게']>80, ['체질량']] = '비만' # '몸무게' 컬럼의 데이터에서 80 이상인 행을 추출한 다음, 해당 행에 대해서만 '체질량' 컬럼에 '비만'을 입력
* iloc
  * loc와 형식은 동일한데 loc달리 행과 컬럼의 index 정보로만 접근
