```python
customer_tuples = [
    ("C001", "김철수", 35, "남성",150000,"2024-03-15"),
    ("C002", "이영희", 35, "여성",280000,"2024-03-14"),
    ("C003", "박민수", 35, "남성",95000,"2024-03-13")
]
# 튜플데이터 : (고객ID, 이름, 나이, 성별, 구매금액, 구매날짜)
```

```python
keys1 = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict1 = {key : list(customer) for key, customer in zip(keys1, zip(*customer_tuples))}
print(type(dict1))
dict1
```

```python
import json

json_data = json.dumps(dict1, ensure_ascii = False) # json.dumps : 객체 -> [문자열]
print(type(json_data))
json_data
```

```python
obj_data = json.loads(json_data) # json.loads : 문자열 -> [객체]
print(type(obj_data))
obj_data
```

```python
keys2 = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict2 = [{k:v for k,v in zip(keys2,customer)} for customer in customer_tuples]
dict2
```

```python
keys_id,*keys_info = ("ID","이름", "나이", "성별", "구매금액", "구매날짜")
dict3 = {id : dict(zip(keys_info,customer)) for id, *customer in customer_tuples}
dict3
```
