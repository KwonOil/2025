```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def calculate_star(star_tag):
    star_rate = ''
    p_cnt = star_tag.text.strip()
    for star in [tag_i.get('class',[])[1] for tag_i in star_tag.find_all('i', class_ = 'fa')]:
        if star == 'fa-star':
            star_rate += '★'
        elif star == 'fa-star-half-o':
            star_rate += '☆'
    return star_rate, p_cnt

def show_info(i, name, category, price, star_rate, p_cnt, link, img):
    print(f"{i}번째 상품")
    print(f"이름 : {name}")
    print(f"카테고리 : {category}")
    print(f"가격 : {price:,}원")
    print(f"평점 : {star_rate} {p_cnt}")
    print(f"상세보기 : {link}")
    print(f"이미지 : {img}")
    print()

# 1 ~ 4 페이지까지 반복
for num in range(1,5):
    url = f"https://startcoding.pythonanywhere.com/basic?page={num}"

    # URL 경로의 HTML 내용 가져오기
    response = requests.get(url)
    html_cont = response.text

    # print(html_cont)

    # 파싱
    soup = BeautifulSoup(html_cont, 'html.parser')

    # 각 페이지에서 몇 번째 상품인지 부여
    i = 0
    print("="*40, f"{num} 페이지", "="*40)
    for product in soup.find_all(class_ = 'product'):
        # 상품번호
        i += 1

        # 공통된 img 태그
        img_tags = product.find(class_ = 'product-img')
        # 공통된 body 태그
        body_tags = product.find(class_ = 'product-body')

        # 각 요소에 필요한 태그들
        name_tag = body_tags.find(class_ = 'product-name')
        category_tag = body_tags.find(class_ = "product-category")
        price_tag = body_tags.find(class_ = "product-price")
        star_tag = body_tags.find('div', class_ = "product-rating")
        link = body_tags.find('a')['href']
        linkurl = urljoin(url, link)

        img = img_tags.find('img')['src']
        imgurl = urljoin(url,img)

        # 가격에서 할인 전 가격 태그를 제거
        for del_tag in price_tag.find_all("del"):
            del_tag.extract()
        
        # 각 태그에서 추출한 요소들
        name = name_tag.string
        category = category_tag.string
        price = int(price_tag.get_text(strip=True).replace(',','').replace('원',''))
        star_rate, p_cnt = calculate_star(star_tag)
        
        # 출력
        # if star_rate == 5:
        show_info(i, name, category, price, star_rate, p_cnt, linkurl, imgurl)
```
