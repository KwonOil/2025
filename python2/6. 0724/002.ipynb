```python
columns = ["제품코드","제품명","카테고리","재고수량","가격","입고일"]
inventory_tuple = [
    ("P001", "노트북", "전자기기", 50, 1200000, "2024-02-01"),
    ("P002", "스마트폰", "전자기기", 100, 850000, "2024-02-15"),
    ("P003", "이어폰", "액세서리", 200, 180000, "2024-03-01")
]
```

```python
def toJSON(keys, tuples):
    return [ dict(zip(keys,i)) for i in inventory_tuple]
# result = [ dict(zip(keys,i)) for i in inventory_tuple]
data = toJSON(columns,inventory_tuple)
```

```python
import json
json_data = json.dumps(data, ensure_ascii=False) # json.dumps : 객체를 [문자열로] 변환
print(type(json_data))
print(json_data)

with open("test_write.json","w",encoding="utf-8") as f:
    f.write(json_data)
```

```python
with open("test_write.json","r",encoding="utf-8") as f:
    json_data2 = f.read()

json_obj = json.loads(json_data2) # json.loads : 문자열을 [객체로] 변환
print(type(json_obj))
print(json_obj)
json_obj[0]['제품명']
```

```python
name = input("제품은?")
for item in json_obj:
    if item['제품명'] == name:
        print(f'''{item['제품명']} 단가 : {item['가격']:,}원, 총합계 : {item['가격']*item['재고수량']:,}원''')
```
