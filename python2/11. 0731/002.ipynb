```python
filename = 'students2.txt'

result = {}

with open(filename, 'r', encoding='utf-8') as f:
    header = next(f).strip().split(',')

    for k in header:
        result[k] = []
    for line in f:
        values = line.strip().split(',')
        for k,v in zip(header, values):
            result[k].append(int(v) if v.isdigit() else v)
result
```

```python
def read_students(filename):
    data = {}
    with open(filename, 'r', encoding='utf-8') as f:
        header = next(f).strip().split(',')
        for line in f:
            values = line.strip().split(',')
            for k, v in zip(header, values):
                if k not in data:
                    data[k] = []
                data[k].append(int(v) if v.isdigit() else v)
    return data
filename = 'students2.txt'

data2 = read_students(filename)
data2
```
